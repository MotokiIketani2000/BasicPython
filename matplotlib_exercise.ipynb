```python
# 使うライブラリ
'''
python3.12以降を使用してる場合、japanize-matplotlibは使えません。
その場合、matplotlib-fontjaを使用してください。
'''
# !pip install matplotlib-fontja

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist = pd.read_csv("hist_data.csv")
plt.figure(figsize=(8, 6))
plt.hist(df_hist.iloc[:, 0], bins=10, color="skyblue", edgecolor="black")
plt.title("ヒストグラム")
plt.xlabel("値")
plt.ylabel("度数")
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter = pd.read_csv("scatter_data.csv")
plt.figure(figsize=(8, 6))
plt.scatter(df_scatter["x"], df_scatter["y"], color="coral")
plt.title("散布図")
plt.xlabel("x 軸")
plt.ylabel("y 軸")
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line = pd.read_csv("line_data.csv")
plt.figure(figsize=(8, 6))
plt.plot(df_line["x"], df_line["y1"], marker="o", linestyle="-", color="blue")
plt.title("折れ線グラフ (y1)")
plt.xlabel("x 軸")
plt.ylabel("y1")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df_line["x"], df_line["y2"], marker="s", linestyle="--", color="green")
plt.title("折れ線グラフ (y2)")
plt.xlabel("x 軸")
plt.ylabel("y2")
plt.show()
```
