## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist = pd.read_csv("./data/hist_data.csv")
a_data = df_hist["aデータ"]
b_data = df_hist["bデータ"]
plt.figure(figsize=(6,4))
plt.hist(a_data, bins=20, alpha=0.5, label="aデータ", color="orange")
plt.hist(b_data, bins=20, alpha=0.5, label="bデータ", color="turquoise")
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter = pd.read_csv("./data/scatter_data.csv")
df_0 = df_scatter[df_scatter["class"]==0]
df_1 = df_scatter[df_scatter["class"]==1]
df_2 = df_scatter[df_scatter["class"]==2]
plt.figure(figsize=(8, 6))
plt.scatter(df_0["length (cm)"], df_0["width (cm)"], color="red", label="class0")
plt.scatter(df_1["length (cm)"], df_1["width (cm)"], color="blue", label="class1")
plt.scatter(df_2["length (cm)"], df_2["width (cm)"], color="green", label="class2")
plt.title("クラス0〜2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line = pd.read_csv("./data/line_data.csv")
plt.figure(figsize=(8, 6))
plt.plot(df_line["期間"], df_line["件数"], marker="o", linestyle="-", color="blue")
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df_line["期間"],df_line["費用"], color="steelblue")
plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
